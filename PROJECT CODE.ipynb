```python
# Step 3: Perform PCA.

[m3,n3]=(x_train.shape)
print('total number of train images= ',m3)
#pca = PCA(n_components=150).fit(x_train)
var1=x_train
[m,n]=var1.shape
means=var1.mean(0)
for i in range (m):
     np.array(var1)[i,:]=np.array(var1)[i,:]-np.array(means)
cmat=np.dot((var1.T),var1)
cmat=cmat/(n3-1)
evals,evec=np.linalg.eigh(cmat)
# xx=np.arange(0,4096)
# plt.scatter(xx,evals)
p=int(input('enter the principal component that you want= '))
print('show_eigenfaces')

basis=evec[:,4096-p:]
#show_eigenfaces(basis)



```
